```python
from src import process
import os
import glob
import random
from src import Time

file_path = '.\dat\HY202103\**\*LMZ?.xml'

xml = []
for filename in glob.glob(file_path, recursive=True):
    xml.append(filename)
    
i = random.randint(0,98)

def function(i):
    if 41< i <56 :
        process.fitting(xml[i],1,1,Time.time())
        print('This fitting figure was not well fitted')

    elif 69< i <84 :
        process.fitting(xml[i],1,1,Time.time())
        print('This fitting figure was not well fitted')

    else:
        process.fitting(xml[i],1,1,Time.time())
        print('This fitting figure was measured well with I-V Data.')


if 41< i <56 :
    a = random.randint(0,98)
    function(a)
    if 41<a<56:
        b = random.randint(0,98)
        function(b)
    elif 69< a < 84:
        b = random.randint(0,98)
        function(b)
    else:
        process.fitting(xml[a],1,1)
        print('This fitting figure was measured well with I-V Data.')

elif 69< i <84:
    a = random.randint(0,98)
    function(a)
    if 41<a<56:
        b = random.randint(0,98)
    elif 69< a < 84:
        b = random.randint(0,98)
        function(b)
    else:
        process.fitting(xml[a],1,1,Time.time())
        print('This fitting figure was measured well with I-V Data.')

else:
    process.fitting(xml[i],1,1,Time.time())
    print('This fitting figure was measured well with I-V Data.')
```

## 6. 'Max transmission of Ref. spec.'

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import glob

csv_list = glob.glob("./res/csv/*.csv")
csv_list = csv_list[-1]
df_res = pd.read_csv(csv_list)

ax = plt.subplot(2,2,1)
g = plt.gcf()
g.set_size_inches(12, 12)  #그림 size 수정_JH
g = sns.scatterplot(data=df_res,  x="Wafer", y="Max transmission of Ref. spec. (dB)",hue ="Analysis Wavelength",palette ="deep", ax = ax )
ax = plt.subplot(2,2,2)
g=sns.scatterplot(data=df_res, x = "Wafer", y = "Rsq of IV")
ax = plt.subplot(2,2,3)
h = sns.scatterplot(data=df_res, x = "Wafer", y = "I at 1V [A]", hue = "Analysis Wavelength", palette = "colorblind", ax = ax)
ax = plt.subplot(2,2,4)
i = sns.scatterplot(data=df_res, x = "Wafer", y = "I at -1V [A]", hue = "Analysis Wavelength", palette = "colorblind", ax = ax)
```

### 1) scatter
 - Orange is 1550nm wavelength and blue is 1310nm wavelength.

### 2) Standard deviation
* Max transmission of Ref. spec. shows the mean, variance, and standard deviation for each wafer and wavelength.
